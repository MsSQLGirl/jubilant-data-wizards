## Step 4. Install Kqlmagic

There are two options in installing Kqlmagic:

**Option 1**. Install Kqlmagic through **_Manage Packages_** dialog in aÂ **_Python_** Notebook.Â 

> Note: this button only appears in a Notebook connected to **_Python 3_** kernel.

  

Search for Kqlmagic and install **the latest non-dev version**.Â 



**Option 2**: Use the following script to install Kqlmagic via Python script

```python
import sys, os

# Upgrade pip first
cmd = f'{sys.executable} -m pip install --upgrade pip'
cmdOutput = !{cmd}
print (cmdOutput)

# Upgrade Kqlmagic to the latest non-dev version
kqlmagicInstall = 'Kqlmagic --no-cache-dir --upgrade'
cmd = f'{sys.executable} -m pip install {kqlmagicInstall}'
cmdOutput = !{cmd}
# print(cmdOutput)

# Check Kqlmagic installation
cmd = f'{sys.executable} -m pip show Kqlmagic'
cmdOutput = !{cmd}
# print (cmdOutput)

if len(cmdOutput) > 0 and '0.1.113' in cmdOutput[1]: 
    print('Success: Kqlmagic required version is already installed!')
else:
    print(f'\nError. There was an issue with installing Kqlmagic')

```
